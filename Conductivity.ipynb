### Quality Analysis and Assignment of weights

As discussed in Section 2.5 of the cited meta-analysis, a Quality Assessment Checklist was administered for each study using a 14 item list. A scaled numerical value was assigned according to the studies compliance with each item. The sum, divided by the number of relevant items was calculated to provide a final Quality Assessment Score (QAS). 

A relative error for each conductivity value was calculated as the standard deviation percentage of a multitude of values for a single tissue type for each participant (if the method is 100% precise, each value for the same tissue should be the same) or the error attributed to the measurement protocol – both described as a decimal. If both the standard deviation and measurement error were provided, the standard deviation was used to calculate the relative error. The relative error was then subtracted from one (where one indicates complete confidence the conductivity value is 100% accurate) to obtain a final confidence value of which the maximum is one.

The QAS of each study and the relative error of each conductivity value were combined to provide a “weight”. This weight was calculated by multiplying the QAS by the confidence value (both with a maximum of one). The maximum associated weight each value has towards the analysis is therefore one. Values assigned weights closer to one were therefore regarded as being more accurate.

The Quality Assessment Protocol and assigned QAS for each study is provided below. The "weights" for each conductivity measurement are provided in the respective weight column in the above table.
