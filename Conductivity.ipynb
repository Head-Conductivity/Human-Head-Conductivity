## Variation in Head Tissue Conductivity Values
A summary of all available electrical conductivity (S/m) values for different head tissues is provided based on on the paper [Variation in Reported Human Head Tissue Electrical Conductivity Values](https://link.springer.com/article/10.1007/s10548-019-00710-2). Results and corresponding figures are updated when new data is available. If you become aware of any new data, please do not hesitate to contact us.
